###  Handle Outliers

```python
# Detect and handle outliers (example for price column)
```

```python
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1  # Interquartile range
# Remove rows with price outliers
df = df[~((df['Price'] < (Q1 - 1.5 * IQR)) | (df['Price'] > (Q3 + 1.5 * IQR)))]
df['Price']
df
```

```python
# Visualize outliers in the car price column
sns.boxplot(df['Price'])
plt.show()
```

```python
# Standardize Values: Make sure values in categorical columns (like "Fuel Type" or "Brand") are consistent (e.g., no multiple spellings of the same brand).
```

```python
df['Brand']=df['Brand'].str.title()
```

```python
df['Brand']
```

```python
df['Fuel'] = df['Fuel'].str.lower()  # Convert fuel types to lowercase
df['Fuel'] 
```

```python
df['Gearbox'] = df['Gearbox'].str.lower().replace({'auto': 'automatic'})  # Standardize transmission types
df
```

```python
# After cleaning, inspect the data again
(df.info())
```

# Exploratory Data Analysis (EDA)

### A. Univariate Analysis:¶

Analyze the distribution of individual variables like Car Price, Car Brand, Gearbox, and Fuel Type.

### Brand Distribution

Analyze the distribution of car brands in Australia. Which car brands are the most common? Are there any brands that dominate the market?


```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Brand'])
plt.title('Brand Distribution')
plt.xlabel('Brands')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
# Use Horizontal Bar Plot
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(y='Brand', data=df)
plt.title('Brand Distribution')
plt.xlabel('Brands')
plt.ylabel('Frequency')
plt.show()
```

 ***Limit Number of Categories(for top 20 cars)***

```python
top_brands = df['Brand'].value_counts().nlargest(20).index
plt.figure(figsize=(12, 6))
sns.histplot(df[df['Brand'].isin(top_brands)]['Brand'])
plt.xticks(rotation=90)
plt.show()
```

### Conclusion:

**Toyota** is the most common car brand in the dataset, indicating that it dominates the Australian car market. This could be due to Toyota’s reputation for reliability, affordability, and popularity among Australian consumers.

```python
# Visualize car fuel distribution
sns.countplot(df['Fuel'])
plt.xlabel('Frequency')
plt.ylabel('Fuel Type')
plt.show()
```

# Price Analysis

Investigate the pricing trends of cars. Are certain brands more expensive than others? What is the average price for each brand, and how does 

```python
# Group by Brand and Calculate Average Price
```

```python
brand_avg_price =df.groupby(['Brand'])['Price'].mean().sort_values(ascending=False)
brand_avg_price 
```

```python
# Visualize the Data with Bar Plots
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=brand_avg_price.index, y=brand_avg_price.values)
plt.xticks(rotation=90)
plt.title('Average Car Price by Brand')
plt.xlabel('Brand Name')
plt.ylabel('Average Price')
plt.show()
```

```python
# Analyze the Price Distribution for Each Brand using boxplot
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['Brand'],y=df['Price'])
plt.xticks(rotation=90)
plt.title('Average Car Price by Brand')
plt.xlabel('Brand Name')
plt.ylabel('Average Price')
plt.show()
```


## Top 5 Most Expensive Cars

Which are the 5 most expensive cars in the dataset? What are their brands, models, and features?


```python
top_5_expensive_cars=df.sort_values(by="Price",ascending=False).head()
top_5_expensive_cars
```

### Inspect the Top 5 Cars

```python
# Display the top 5 most expensive cars with important columns
top_5_expensive_cars[['Brand', 'Model', 'Price', 'Year','Variant', 'CC','Seating Capacity']]
```

### Price Analysis 

```python
# Visualize the Top 5 expensive cars
```

```python
# Plot a bar chart for the Top 5 expensive cars
plt.figure(figsize=(8, 4))
sns.barplot(x='Brand', y='Price', data=top_5_expensive_cars)
plt.title('Top 5 Most Expensive Cars')
plt.xlabel('Car Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

### Which brands have the highest average prices?

```python
brand_avg_prices = top_5_expensive_cars.groupby('Brand')['Price'].mean().sort_values(ascending=False)
brand_avg_prices
```

```python
# Create a bar plot for visualization
plt.figure(figsize=(10, 6))  # Set figure size
sns.barplot(x=brand_avg_prices.index, y=brand_avg_prices.values)
plt.title('Top Brands with Highest Average Prices')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.show()
```

```python
# Scatter plot for average prices by brand
plt.figure(figsize=(10, 6))
sns.scatterplot(x=brand_avg_prices.index, y=brand_avg_prices.values)
plt.title('Top Brands with Highest Average Prices')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.show()
```

### Features Analysis

```python
# Plot a bar chart for the Top 5 expensive cars
plt.figure(figsize=(8, 4))
sns.barplot(x='Brand', y='CC', data=top_5_expensive_cars)
plt.title('Top 5 Most Expensive Cars')
plt.xlabel('Car Model')
plt.ylabel('CC')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot a bar chart for the Top 5 expensive cars
plt.figure(figsize=(8, 4))
sns.barplot(x='Brand', y='Seating Capacity', data=top_5_expensive_cars)
plt.title('Top 5 Most Expensive Cars')
plt.xlabel('Car Model')
plt.ylabel('Seating Capacity')
plt.show()
```

# Conclusion:

**Ford** has the highest average price among all brands, indicating that it is priced higher on average compared to other car brands in the dataset.
Ford also has the maximum engine capacity (CC), making it stand out in terms of performance-oriented features.

**Mercedes-Benz** has the highest seating capacity, indicating that this brand may cater more to larger vehicles like SUVs or luxury sedans with more passenger space.

This highlights that **Ford** dominates in engine performance, while Mercedes-Benz focuses on comfort and space..
